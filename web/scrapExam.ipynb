```python

import requests as req
from bs4 import BeautifulSoup # html 해석기
response = req.get("https://www.naver.com/")


html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""
# print(html_doc)

soup = BeautifulSoup(html_doc, 'html.parser')
# soup를 이용해서 html_doc 문자열 파싱, 그 내용을 Soup 변수에 대입

a_list = soup.select('a')
print(a_list)
print('a태그 리스트 중 첫번째 텍스트')
print(a_list[0])

print(a_list[0].text) # 첫번째 a에서 텍스트 가져오기

for a in a_list:
    print(a.text) # a태그 리스트에 있는 텍스트 모두 출력

for link in a_list: # a태그 리스트에 있는 링크 모두 출력
    print(link['href']) # 속성을 선택할 때는 대괄호를 사용

```

```python
# html 선택방법 (css 선택자)
# 1. 태그 선택
# 2. 후손 선택
# 3. 자식 선택
# 4. class 선택
# 5. id 선택

# print(html_doc)

soup = BeautifulSoup(html_doc, 'html.parser')
# soup을 이용해서 html_doc 문자열 파싱, 그 내용을 soup 변수에 대입

a_list = soup.select('a')

print("== a태그들 리스트 ==")
print(a_list)

print("== a태그 리스트 중 첫 번째 ==")
print(a_list[0])

print("== a태그 리스트 중 첫 번째 텍스트 ==")
print(a_list[0].text)

print("== a태그 리스트 중 첫 번째 링크 ==")
print(a_list[0]['href']) # 속성을 선택할 땐 대괄호

print("== a태그 리스트에 있는 텍스트 모두 출력 ==")
for a in a_list:
    print(a.text)

print("== a태그 리스트에 있는 링크 모두 출력 ==")
for link in a_list:
    print(link['href'])


```

```python
import requests as req
from bs4 import BeautifulSoup 

response = req.get("https://news.naver.com/")
print(response)
print(response.text)

# html 문자열 파싱
soup = BeautifulSoup(response.text, 'html.parser')

content = soup.select('.cnf_news_title')
print(content[0].text)

for c in content:
    print(c.text)


content2 = soup.select('.cnf_title_area > h4')
for c in content2:
    print(c.text)


imgs = soup.select('.cnf_news_thumb > img')
for img in imgs:
    print(img) # print(img)
```
