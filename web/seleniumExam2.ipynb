```python
# 요소 찾는 방법 3가지
# 1. 태그 이름(TAG NAME)
# 2. class / id(주민등록번호) (CLASS_NAME / ID)
# 3. xPath

from selenium import webdriver
from selenium.webdriver.common.by import By


#===============================
# 0) 브라우져 실행 + 페이지 접속
#===============================

driver = webdriver.Chrome()
driver.get("https://quotes.toscrape.com/")

# 수집한 결과를 담을 리스트
results = []


#=====================================
# 1) 태그 이름(TAG_NAME)으로 요소 찾기
#=====================================

# - HTML 태그(div, span, a등)로 요소 찾기
# - 범위가 너무 넓어지는 경우가 많아서 실무에서는 주로 '연습용'으로 사용

div_elements = driver.find_elements(By.TAG_NAME, 'div')
print('TAG_NAME으로 찾은 div 태그 개수 :', len(div_elements))


#=====================================
# 2) class로 요소 찾기 (CLASS_NAME)
#=====================================

# HTML에서 class='text'처럼 지정된 class 이름으로 요소 찾기
# Selenium에서 가장 자주 사용하는 방식 중 하나

# 주의 : 
# - class에 공백이 포함된 경우(예 : class='a b)는
# CLASS_NAME으로 바로 찾기 어려움

first_quote_text = driver.find_element(By.CLASS_NAME, 'text')
print('\nCLASS_NAME으로 찾은 첫 번째 명언 :,', first_quote_text.text)


#=====================================
# 3) id로 요소 찾기 (ID)
#=====================================

# id는 보통 페이지에서 "유일"하게 존재
# 그래서 정확하게 찾고 싶을때 가장 안정적

# quotes_container = driver.find_element(By.ID, "quotes")
# print('\nID로 찾은 요소의 태그이름 :',quotes_container.tag_name)
# 현재는 ID가 없어서 찾지 못하고  에러발생


#=====================================
# 4) find_element VS find_elements
#=====================================
 
# find_element : 요소 1개만 찾음(없으면 에러발생) 
# find_elements : 요소 여러개를 찾고 리스트로 변환(없으면 빈리스트)

one_quote_block = driver.find_element(By.CLASS_NAME, 'quote')
many_quote_blocks = driver.find_elements(By.CLASS_NAME, 'quote')

print('\nfine_element로 찾은 quote 블록 1개 :', type(one_quote_block))
print('\nfine_elements로 찾은 quote 블록 개수 :', len(many_quote_blocks))


#=====================================
# 5) xPath로 요소 찾기
#=====================================

# - HTML 구조(경로)를 기준으로 요소를 찾는 방식
# - class, id가 애매하거나 조건이 복잡할 때 유용

# 장점 - 조건/경로 기반으로 정확하게 찾기 쉬움
# 단점 - HTML 구조가 바뀌면  xPath 도 함께 깨질 수 있음

# 첫 번재 명언 텍스트를  xPath로 찾기

xPath_quote_text = driver.find_element(
    By.XPATH,
    '//div[@class='quote']/span[@class='text']'
)

# -// : 문서 전체에서 검색(root부터 전역 탐색)
# -div[@class='quote'] : div 태그 중 class='quote'인 요소
# - / : 바로 아래 자식(child) 관계
# - span[@class='text'] : 위에서 찾은 div 바로 아래에 있는 span 중에서 class

# 의미 - class='quote'인 div안의 class_'text'인 span을 찾아라

print('\nxPath로 찾은 첫번째 명언:', xPath_quote_text.text)


# 첫번째 quote 블록만 특정해서 찾는 xPath

xPath_first_block = driver.find_element(By.XPATH,'(\\div[@class='quote'])')
xPath_first_author = xPath_first_block.find_elememt(By.CLASS_NAME,'author')

print('XPath로 찾은 첫번재 저자:', xPath_first_author.text)










```
