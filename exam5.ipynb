```python
# 정규표현식(정규식)
# 정규식을 쓸때는 r을 붙여야 한다

text = '안녕하세요 저는 15살 입니다.'
print('15' in text)
'15' in text

import re
text = '안녕하세요 저는 15살 입니다. 저는 안경이 15개 있습니다'
pattern = r'\d+' # r은 raw string \(역슬래쉬는 특수문자)

result = re.findall(pattern, text)
print(result)
```

```python
sample = 'abc123def456'

re.findall(r'\d+', sample)

re.findall(r'[a-z]+', sample)
```

```python
# 특정 형식 찾기

data = 'test123@gmail.com'

email_pattern = 'r\w+@\w+\.\w+'
# \w : 문자 또는 숫자
# + : 1개 이상
# @ : 말 그대로 @
#\. : .

emails = re.findall(email_pattern, data)
print(emails)
```

```python
# 문자열 치환

test = '가격은 1200원 입니다.'

new_text = re.sub(r'\d+','',test)
print(new_text)
```

```python
# 정규식과 일반 문자열 처리

test = '몸무게: 80kg, 키: 180cm'
test.split(',')
numbers = re.findall(r'\d+', test)
print(numbers)
```
