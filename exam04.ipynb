```python
# 함수(자판기) 
# -매개변수(자판기에 있는 동전(데이터)투입구) -리턴(자반기에 있는 음료수(데이터)배출구)
# 함수를 사용해서 코드량을 확 줄여주세요.
# 자주 사용하는 코드를 함수로 만들어 두면 참 편하다
# 이름이 'print8dan'인 함수를 정의한다
def print8Dan():
    dan = 8
    i = 1
    while i <= 9:
        print('{}*{} = {}'.format(dan, i, dan*i))
        i += 1
print8Dan()
```

```python
# 문제 : 입력받은 정수가 소수인지 아닌지 알려주는 함수를 구현해주세요.

def is_prime_number(num):
  if num == 1:
    return True

  i = 2
  while i < num:
    if num % i == 0:
      return False
    i += 1
  return True
    

print("3은 소수입니다 : {}".format(is_prime_number(3)))
print("4는 소수입니다 : {}".format(is_prime_number(4)))
print("5는 소수입니다 : {}".format(is_prime_number(5)))
print("6은 소수입니다 : {}".format(is_prime_number(6)))
print("7은 소수입니다 : {}".format(is_prime_number(7)))
print("1000은 소수입니다 : {}".format(is_prime_number(1000)))

    

```
