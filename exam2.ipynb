```python
# 파이썬 유니코드

text = '파이썬'

print(text)

print(len(text))

# 2. 문자, 숫자 변환
print(ord('가'))
```

```python
# 2. 문자, 숫자 변환
print(ord('가'))
print(chr(44032))
```

```python
# 3. 문자열을 바이트로 변환(encode)
# -encode : 문자열을 바이트로 바꾸는 작업

encoded = text.encode('utf-8')
print(encoded)
print(type(encoded))
```

```python
# 4. 바이트를 다시 문자열로

decoded = encoded.decode('utf-8')
print(decoded)
print(type(decoded))
```
