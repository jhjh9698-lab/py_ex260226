```python
# 클래스 - 객체를 표현하기 위한 문법의 일종, 
# 프로그래밍으로 객체를 만들때 사용하는것

# 게임 캐릭터도 객체이다 - 전사가 갑옷을 입고 있고 칼을 차고 있고 머리가 검정이다 
# 전사, 마법사, 궁수, 자동차, 비행기 => object
# => 현실, 가상, 개념

# 속성(명사적) - attribute : 머리 모양, 머리 색깔 
# 능력(기능, 동사적) - method : 공격하다, 방어하다


class person:
    def hello(self): #self는 꼭 써야 한다(법칙)
        print('hi')

jin = person() #jin은 person의 인스턴스가 된다
jin.hello()
```

```python
# 클래스 기본 사용법

class person:
    def hello(self):
        print('hello')

    def hi(self):
        self.hello()    

jin = person()
jin.hi()
```

```python
# 인스턴스 속성 - 언더바 2개사용

class person():
    def __init__(self):
        self.hello = '안녕하세요' # 인스텬스 속성

    def hi(self):
        print(self.hello)   
jin = person() # 이때 __init__메서드는 이미 실행됐다
jin.hi()         
```

```python
class Person():
  def __init__(self, name, age, hobby): # 이 안에 있는 속성은 인스턴스 속성이다.
    self.hello = '안녕하세요'
    self.name = name
    self.age = age
    self.hobby = hobby

  def introduce(self):
    print("{}, 저는 {} 입니다.".format(self.hello,self.name)) # 클래스 내부에서 접근

jin = Person('진', 22, '운동')
jin.introduce()

paul = Person('폴', 33, '요리')

print("== 진 ==")
print('이름 : {}'.format(jin.name)) # 클래스 외부에서 접근
print('나이 : {}'.format(jin.age))
print('취미 : {}'.format(jin.hobby))

print("== 폴 ==")
print('이름 : {}'.format(paul.name))
print('나이 : {}'.format(paul.age))
print('취미 : {}'.format(paul.hobby))
```

```python
# 클래스 외부에서는 접근불가, 클래스 내부에서만 사용가능한 private attribute
class Person():
  def __init__(self, name, age, hobby, weight):
    self.hello = '안녕하세요'
    self.name = name
    self.age = age
    self.hobby = hobby
    self.__weight = weight # 변수 앞에 __ 를 붙이면 비공개 속성으로 변경

  def diet(self, loss_weight):
    self.__weight -= loss_weight
    print("{} kg 입니다.".format(self.__weight))

jin = Person('진', 22, '운동', 78)
jin.diet(20)
```

```python
# 인스턴스 속성

class Person():
  def __Hello(self):
    print('hello')

  def hi(self):
    self.__Hello()

jin = Person()
jin.hi()
jin.__Hello() # 비공개 메서드라서 호출이 안된다.
```
