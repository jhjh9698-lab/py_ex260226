```python
# 클래스 - 객체를 표현하기 위한 문법의 일종, 
# 프로그래밍으로 객체를 만들때 사용하는것

# 게임 캐릭터도 객체이다 - 전사가 갑옷을 입고 있고 칼을 차고 있고 머리가 검정이다 
# 전사, 마법사, 궁수, 자동차, 비행기 => object
# => 현실, 가상, 개념

# 속성(명사적) - attribute : 머리 모양, 머리 색깔 
# 능력(기능, 동사적) - method : 공격하다, 방어하다


class person:
    def hello(self): #self는 꼭 써야 한다(법칙)
        print('hi')

jin = person() #jin은 person의 인스턴스가 된다
jin.hello()
```

```python
# 인스턴스 속성 - 언더바 2개사용

class person():
    def __init__(self):
        self.hello = '안녕하세요' # 인스텬스 속성

    def hi(self):
        print(self.hello)   
jin = person() # 이때 __init__메서드는 이미 실행됐다
jin.hi()         
```
