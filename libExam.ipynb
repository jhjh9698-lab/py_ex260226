```python
# 데이터 형식 모듈
# json 형식으로 데이터 변환
# json - 데이터를 약속된 모양으로 적는 방식('key','value')
# 웹, API, 파일 저장에 사용
# API - 프로그램기리 대화하는 방법(규칙)

import json
data = {'name' : 'jin', 'age' : 20}
print(json.dumps(data))
```
